```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from matplotlib import colors
```

```python
df1 = pd.read_csv("csv/LJdata_eq1.csv").sort_values("T")
df2 = pd.read_csv("csv/LJdata_eq2.csv").sort_values("T")
```

```python
df1
```

```python
df2
```

```python
fig, ax = plt.subplots(1,1,figsize=(3.5,3))

plt.scatter( df2["T"], df2["U"] , label="LJ")

plt.ylabel(r"U(T)")
plt.xlabel("Temperature (K)")
plt.title("Our Results \n 795 atom Graphene Sheet \n Using eq2")
plt.legend()
plt.savefig(f'plots/LJ_U_vs_T.png', dpi=150, bbox_inches='tight')
```

```python
kcal_to_J = 4184

fig, ax = plt.subplots(1,1,figsize=(3.5,3))

plt.scatter( df2["T"][1:], df2["C_v_e"][1:]*kcal_to_J , label="LJ")

plt.ylabel(r"$C_V$ ( J/(mol*K) )")
plt.xlabel("Temperature (K)")
plt.title("Our Results \n 795 atom Graphene Sheet \n Using eq2")
plt.legend()
plt.savefig(f'plots/LJ_C_v_eq2.png', dpi=150, bbox_inches='tight')
```

```python
kcal_to_J = 4184

fig, ax = plt.subplots(1,1,figsize=(3.5,3))

plt.scatter( df1["T"][1:], df1["C_v_e"][1:]*kcal_to_J , label="LJ")

plt.ylabel(r"$C_V$ ( J/(mol*K) )")
plt.xlabel("Temperature (K)")
plt.title("Our Results \n 795 atom Graphene Sheet \n Using eq1")
plt.legend()
plt.savefig(f'plots/LJ_C_v_eq1.png', dpi=150, bbox_inches='tight')
```
